# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import cm
```

# Reading Dataset

```python
ds = pd.read_csv('../datasets/dataset_clean.csv')
ds.head()
```

# Data Visualization & Analysis

--------------------------------------------------------------------------------------------------------------------
The figure in cell 3 represents the **time seires analysis** (3rd April 2018 to 17th October 2019) for each genre of apps available in the Apple App Store.

- Average Price (USD) of **Medical Apps** was highest around **May 2019**. 
- Average Price (USD) of **Reference Based Apps** was highest around **September 2019**.
- Average Price (USD) of **Navigational Apps** was highest around **July 2018**.
- Average Price (USD) of **Productivity Based Apps** was highest around **October 2019**.

Based on the trends displayed, one could identify the appropriate time period to release his/her application (when it is most in demand) in order to maximize the profits.

--------------------------------------------------------------------------------------------------------------------

```python
# Dataset for Medical Time Series Analysis
ds_med = ds.loc[ds['Primary_Genre'] == 'Medical']
ds_med = ds_med[['Original_Release_Date', 'Price_USD']]
ds_med = ds_med.loc[(ds_med['Original_Release_Date'] >= '2018-04-03')]
ds_med = ds_med.drop_duplicates()
ds_med = ds_med.sort_values(by='Original_Release_Date').reset_index(drop=True)

# Dataset for Reference Time Series Analysis
ds_ref = ds.loc[ds['Primary_Genre'] == 'Reference']
ds_ref = ds_ref[['Original_Release_Date', 'Price_USD']]
ds_ref = ds_ref.loc[(ds_ref['Original_Release_Date'] >= '2018-04-03')]
ds_ref = ds_ref.drop_duplicates()
ds_ref = ds_ref.sort_values(by='Original_Release_Date').reset_index(drop=True)

# Dataset for Navigation Time Series Analysis
ds_nav = ds.loc[ds['Primary_Genre'] == 'Navigation']
ds_nav = ds_nav[['Original_Release_Date', 'Price_USD']]
ds_nav = ds_nav.loc[(ds_nav['Original_Release_Date'] >= '2018-04-03')]
ds_nav = ds_nav.drop_duplicates()
ds_nav = ds_nav.sort_values(by='Original_Release_Date').reset_index(drop=True)

# Dataset for Productivity Time Series Analysis
ds_pro = ds.loc[ds['Primary_Genre'] == 'Productivity']
ds_pro = ds_pro[['Original_Release_Date', 'Price_USD']]
ds_pro = ds_pro.loc[(ds_pro['Original_Release_Date'] >= '2018-04-03')]
ds_pro = ds_pro.drop_duplicates()
ds_pro = ds_pro.sort_values(by='Original_Release_Date').reset_index(drop=True)

# Creating Subplots and initializing the figure size.
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize=(18, 12))

# Heading
fig.suptitle('Blue = Medical | Red = Reference | Green = Navigation | Black = Productivity', fontsize=16)

# Plots
ds_med.plot(x = 'Original_Release_Date', y = 'Price_USD', kind = 'line', ax = ax1, color='blue')
ds_ref.plot(x = 'Original_Release_Date', y = 'Price_USD', kind = 'line', ax = ax2, color='red')
ds_nav.plot(x = 'Original_Release_Date', y = 'Price_USD', kind = 'line', ax = ax3, color='green')
ds_pro.plot(x = 'Original_Release_Date', y = 'Price_USD', kind = 'line', ax = ax4, color='black')

# Rotating X Axis.
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)
ax3.tick_params(axis='x', rotation=45)
ax4.tick_params(axis='x', rotation=45)
```
