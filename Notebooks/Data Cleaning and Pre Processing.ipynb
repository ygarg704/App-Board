# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Reading Dataset

The original dataset is not available in this repository (because of the size constraints). To run the cell below, download the original dataset from [here](https://www.kaggle.com/cmqub19/763k-ios-app-info)

```python
ds = pd.read_csv('../datasets/dataset.csv')
ds.head()
```

# Data Cleaning and Pre-Processing

```python
# Dropping irrelevant columns.
ds.drop(ds.columns[[0, 1, 2, 4, 5, 6, 16, 18]], axis = 1, inplace = True)
ds.head()
```

```python
ds.info()
```

```python
# Removing Null Values
ds = ds.dropna()
```

```python
ds.info()
```

```python
# Date Time Conversion 
ds['Original_Release_Date'] = pd.to_datetime(ds['Original_Release_Date']).dt.strftime('%Y-%m-%d')
ds['Current_Version_Release_Date'] = pd.to_datetime(ds['Current_Version_Release_Date']).dt.strftime('%Y-%m-%d')
ds = ds.drop_duplicates()
ds.head()
```

```python
# Saving Clean Dataset
ds.to_csv('../datasets/dataset_clean.csv', index=False)
```
